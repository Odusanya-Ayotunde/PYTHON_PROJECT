# In dictionary, using the get to access or compute an array

```python
array =  [3,5, "timi", 3, "jane", 5, 4, 3, 3, 3, "timi"]
dict_arr = {}
for i in array:
    dict_arr[i] = dict_arr.get(i,0) +1
print(dict_arr)
```

```python
array =  [3,5, "timi", 3, "jane", 5, 4, 3, 3, 3, "timi"]
dict_arr = {}
for i in array:
    dict_arr[i] = dict_arr.get(i,0) +1
print(dict_arr)
```

```python
counts = []
for i, j in dict_arr.items():
    counts.append(j)
print(counts)
```

```python
d_max = counts[0]
for i in counts:
    if i > d_max:
        d_max = i
print(d_max)
```

```python
for k, n in dict_arr.items():
    if n == d_max:
        print(k)
```

# using a while loop to build a guess game

```python
guess_here = 10
guess_count = 0
number_guess = 3

while guess_count < number_guess:
    guess = int(input("enter guess: "))
    guess_count = guess_count + 1
    if guess_here == guess:
        print("your guess was right")
    else:
        print("guess again!")
else:
        print("game over!!!")
```
